Dans cette partie on va chercher des pistes pas très évidents dans les données en utilisant l'approche _Exceptional Model Mining_. Les algorithmes employés ici sont notamment basé sur [cet article](https://github.com/MarchesLearning/IdeesBrouillonCommunications/blob/master/ExceptionalModelMining.pdf).  
  
On importe d'abord les librairies et les données, fonctions qu'il faut.

```python
import pickle
import numpy as np
import pandas as pd

from sklearn import linear_model

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

complete_preprocessed = pd.read_pickle("preprocessed")
complete = pd.read_pickle("complete")
tsne = pickle.load(open("tSNE", "rb"))
```

On se rappèle que les sujets sont issues d'une quarantaine de pays d'origine. A noter que les pays-bas ne contiennent qu'une seule observation. 

```python
complete.nativeCountry.unique()
```

Maintenant on veut s'intéresser au niveau moyen de la formation par pays. Les ressortissants de quels pays d'origine font partie des "plus cultivés" ? Biensûr c'est un résultat relativement subjectif puisqu'on a adopté l'échelle exponentielle qui favorise les diplômes de haut niveau.  

```python
dictEducation = {} 


for Country in complete.nativeCountry.dropna().unique():
    dictEducation.update({Country: complete_preprocessed.loc[complete.nativeCountry == 
                                                   Country].educationGeneral.mean()})
dictEducation = dict(sorted(dictEducation.items(), key = lambda item: item[1], reverse=True))

plt.figure(figsize = (15, 20))
sns.barplot(y = list(dictEducation.keys()), x = list(dictEducation.values()), orient = "h",
           palette="Blues_d")
plt.show()
```

On peut aussi s'intéresser à une autre question : les salariés de quels pays d'origine gagnent le plus de salaire en moyenne ? Le résultat se calcule en proportion des salaire au-dessus de 50 mille dollars. On voit que l'ordre des salariés n'est pas forcément le même que celui des diplômés.

```python
dictSalary = {} 

for Country in complete.nativeCountry.unique():
    dictSalary.update({Country: complete_preprocessed.loc[complete.nativeCountry == 
                                                   Country]['salary_>50K'].mean()})
    
dictSalary = dict(sorted(dictSalary.items(), key = lambda item: item[1], reverse=True))

plt.figure(figsize = (15, 20))
sns.barplot(y = list(dictSalary.keys()), x = list(dictSalary.values()), orient = "h",
           palette="Greens_d")
plt.show()
```

Après avoir vu ces deux statistiques, on peut se poser une question un peu plus profonde : quel est le lien entre les diplômés et les salariés, en d'autres termes _les études de l'enseignement supérieur aux Etats-Unis par exemple sont-elles équi-rentables pour les ressortissants des différents pays d'origine, ou y a-t-il un certain genre d'inéquilibre entre les pays ?_  
  
Pour répondre à cette question, on va faire recours à l'approche d'_Exceptional Model Mining_. On regarde pour tous les sous-groupes qui sont les pays, quelle est la pente du modèle de régression logistique quand on essaie de prédire le niveau de salaire de chaque sujet avec son niveau de formation au sein du sous-groupe de ce pays d'origine. Plus la pente est grande, plus on peut considérer qu'il est rentable de faire ses études aux Etats-Unis pour les ressortissants de ce pays. A noter qu'à la fin, on ne garde que les pays où le taux de bonnes prédictions du modèle dépasse 70% puisque dans le cas contraire le modèle sera suceptibled'être infidèle.  
  
Cette pente peut s'interpréter en tant qu'un certain indicateur de l'égalité sociale.

```python
dictRentabilite = {}
dictAccuracy = {}

for Country in complete.nativeCountry.dropna().unique():
    
    if Country == "Holand-Netherlands":
        continue
    else:
    
        country = complete_preprocessed.loc[complete.nativeCountry == Country]
        logistic = linear_model.LogisticRegression()
        logistic.fit(country.educationGeneral.reshape(-1, 1), country["salary_>50K"])

        countryOpposant = complete_preprocessed.loc[~ (complete.nativeCountry == Country)]
        logistic2 = linear_model.LogisticRegression()
        logistic2.fit(countryOpposant.educationGeneral.reshape(-1, 1), countryOpposant["salary_>50K"])

        if logistic.coef_[0][0] >= 0.7:

            dictRentabilite.update({Country: logistic.coef_[0][0]}) 
            dictAccuracy.update({Country: logistic.score(
                country.educationGeneral.reshape(-1, 1), country["salary_>50K"])})
```

```python
dictRentabilite = dict(sorted(dictRentabilite.items(), key = lambda item: item[1], reverse=True))

plt.figure(figsize = (15, 20))
sns.barplot(y = list(dictRentabilite.keys()), x = list(dictRentabilite.values()), orient = "h",
           palette="Reds_d")
plt.show()
```

Sans aucune surprise, les Etats-unis sont d'une certaine manière "le plus juste" pour les américains, dans le sens que la valeur de leurs diplômes est le mieux réflétée dans leurs salaires.  
  
Au fil du calcul une partie des pays délivre une pente négative, signifiant que les diplômes sont nagativement rentables pur eux. Or le taux de bonnes prédictions n'atteint 70% dans aucun de ces pays, ce qui veut dire que les salaires de ces ressortissants ne dépendent pas de façon évidente de leurs niveaux d'études.

On peut s'intéresser à une autre petite question intéressante : comment varie-le nombre d'heures de travail hebdomataire selon l'âge du salarié par rapport aux differentes professions ? On emploie ici une deuxième variante de l'approche EMM. Pour chaque profession, on regarde la corrélation entre l'âge du salarié et le nombre d'heure de travail de ce sous-groupe. Le résultat est comme suit. 

```python
corrHoursSalary = {}


for Occupation in complete.occupation.dropna().unique():
    occupation = complete_preprocessed.loc[complete.occupation == Occupation]
    corrHoursSalary.update({Occupation: occupation.age.corr(occupation.hoursPerWeek)})
    
corrHoursSalary = dict(sorted(corrHoursSalary.items(), key = lambda item: item[1], reverse=True))

plt.figure(figsize = (15, 10))
sns.barplot(y = list(corrHoursSalary.keys()), x = list(corrHoursSalary.values()), orient = "h",
           palette="Purples_d")
plt.show()
```

On voit que dans les forces de l'ordre, plus on est âgé, plus on travaille en moyenne. Au contraire pour les gardes, le nombre d'heures de travail diminue lorque l'âge augemente. Pour les métiers comme paysant ou technicien, l'âge n'a peut d'impact sur le nombre d'heures de travail.
